```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
time = np.arange(0, 10, 0.1)
f = np.sin(time)
d = np.cos(time)
d2 = -np.sin(time)
```

```python
plot.plot(time, f, label= "f(x)")
plot.plot(time, d, label= "f'(x)")
plot.plot(time, d2, label= "f''(x)")

# Give a title for the sine wave plot
plot.title("Pontos de inflexão")

# Give x axis label for the sine wave plot
plot.xlabel('x')

# Give y axis label for the sine wave plot
# plot.ylabel('f(x) = sin(time)')

plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.legend()
 

plot.show()
```
